# *IMPORTING DATETIME FOR CLEANING DATE**

```python
from datetime import datetime
```

# Data Cleaning

```python
nation_level_data
```

```python
nation_level_data=nation_level_data.sort_values('Date')
```

```python
nation_level_data
```

# **convert the 'Date' column from "dd monthname" to datetime type in 'dd/mm' format**

### for sorting via date

```python
# convert the 'Date' column to datetime type in 'dd/mm' format
def convert_date(date_string):
    try:
        return datetime.strptime(date_string.strip(), '%d %B').strftime('%d/%m')
    except ValueError:
        return '01/01'
     
```

```python
nation_level_data['Date'] = nation_level_data['Date'].apply(convert_date)
```

```python
nation_level_data
```

# sort the DataFrame by date

```python

# sort the DataFrame by date
nation_level_data = nation_level_data.sort_values(by='Date')
```

```python
nation_level_data
```

# **add 2020 to the date string**

```python
nation_level_data['Date'] = nation_level_data['Date'] + '/2020'
```

# **Visualization**

#**analysis question**
#### **What percentage of daily confirmed cases result in death, and is there a pattern or trend in the data?**

```python
import matplotlib.pyplot as plt

# calculate percentage of daily confirmed cases resulting in death
nation_level_data['Percentage of Daily Confirmed Cases Resulting in Death'] = nation_level_data['Daily Deceased'] / nation_level_data['Daily Confirmed'] * 100

# create scatter plot
plt.scatter(nation_level_data['Daily Confirmed'], nation_level_data['Percentage of Daily Confirmed Cases Resulting in Death'],color='yellow',marker='o',edgecolors='green',linewidth=1,alpha=.75)
plt.xlabel('Daily Confirmed Cases')
plt.ylabel('percentage of Daily Confirmed Cases Resulting in Death')
plt.title('Daily Confirmed Cases vs. Percentage of Daily Confirmed Cases Resulting in Death')
plt.grid()
plt.tight_layout()
plt.show()
```

# **analysis question**
####Is there a correlation between the number of daily recovered cases and daily confirmed cases?

```python
import pandas as pd
import seaborn as sns



# Extract the columns for daily confirmed and daily recovered cases
daily_confirmed = nation_level_data['Daily Confirmed']
daily_recovered = nation_level_data['Daily Recovered']

# Create a scatter plot of daily confirmed cases against daily recovered cases
sns.scatterplot(x=daily_confirmed, y=daily_recovered,hue=daily_recovered)



# Calculate the correlation coefficient
corr = daily_confirmed.corr(daily_recovered)

# Print the correlation coefficient
print(f"Correlation coefficient: {corr}")
```
