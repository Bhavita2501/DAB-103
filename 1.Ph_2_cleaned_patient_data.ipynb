# **Visualiztion of PAtient Data**

---



```python
import matplotlib.pyplot as plt

# Group the data by age and status and count the number of occurrences
counts = pat_data.groupby(["Age Bracket", "Current Status"]).size().reset_index(name="Count")

# Filter only the hospitalized patients
hospitalized_counts = counts[counts["Current Status"]=="Hospitalized"]

# Plot the bar chart
plt.bar(hospitalized_counts["Age Bracket"], hospitalized_counts["Count"])
plt.title("Correlation between age and status of hospitalized patients")
plt.xlabel("Age Bracket")
plt.ylabel("Number of patients")
plt.show()
```
