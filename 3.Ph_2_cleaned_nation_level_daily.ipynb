# *IMPORTING DATETIME FOR CLEANING DATE**

```python
from datetime import datetime
```

# Data Cleaning

```python
nation_level_data
```

```python
nation_level_data=nation_level_data.sort_values('Date')
```

```python
nation_level_data
```

# **convert the 'Date' column from "dd monthname" to datetime type in 'dd/mm' format**

### for sorting via date

```python
# convert the 'Date' column to datetime type in 'dd/mm' format
def convert_date(date_string):
    try:
        return datetime.strptime(date_string.strip(), '%d %B').strftime('%d/%m')
    except ValueError:
        return '01/01'
     
```

```python
nation_level_data['Date'] = nation_level_data['Date'].apply(convert_date)
```

```python
nation_level_data
```

# sort the DataFrame by date

```python

# sort the DataFrame by date
nation_level_data = nation_level_data.sort_values(by='Date')
```

```python
nation_level_data
```

# **add 2020 to the date string**

```python
nation_level_data['Date'] = nation_level_data['Date'] + '/2020'
```
