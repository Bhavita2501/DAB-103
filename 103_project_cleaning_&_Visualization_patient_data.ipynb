```python
# # Load the data into a pandas DataFrame
df = pat_data

# Remove rows that contain numbers in the 'Detected City' column
df = df[~df['Detected City'].str.contains('\d', na=False)]

# Group the data by city and count the number of cases in each city
city_counts = df.groupby('Detected City')['Num Cases'].sum().sort_values(ascending=False)

# Create a bar chart of the top 10 cities with the most cases
plt.figure(figsize=(10,5))
sns.barplot(x=city_counts[:10].index, y=city_counts[:10].values, width=0.8)
plt.title('Cases by City')
plt.xlabel('City')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# How does the number of cases vary across different districts within the same state?
# Load the data into a pandas DataFrame
df = pat_data

# Remove rows where state or district is unassigned
df = df[(df['Detected State'] != 'Unassigned') & (df['Detected District'] != 'Unassigned')]

# Group the data by state and district and count the number of cases in each district
state_district_counts = df.groupby(['Detected State', 'Detected District'])['Num Cases'].sum().reset_index()

# Get a list of unique states
states = state_district_counts['Detected State'].unique()

# Set the number of rows and columns for the plot grid
n_rows = (len(states) + 1) // 2
n_cols = 2

# Create the plot grid
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(12, 6 * n_rows))


# Adjust the top margin of the plot grid
fig.subplots_adjust(top=0.95)

# Add a title to the plot
# fig.suptitle('Cases by District and State', fontsize=16)

# Loop through each state and plot its top 10 districts in a separate subplot
for i, state in enumerate(states):
    # Get the data for the current state and sort by number of cases
    state_data = state_district_counts[state_district_counts['Detected State'] == state].sort_values('Num Cases', ascending=False).head(5)
    if not state_data.empty:
        # Calculate the subplot position
        row = i // n_cols
        col = i % n_cols
        ax = axes[row][col]
        # Create the bar chart in the current subplot
        sns.barplot(x='Num Cases', y='Detected District', data=state_data, ax=ax)
        ax.set_title(f'Cases by District in {state}')
        ax.set_xlabel('Number of Cases')
        ax.set_ylabel('District')
        ax.tick_params(axis='y', labelsize=8)
        ax.tick_params(axis='x', labelsize=8)
        ax.xaxis.grid(True, linestyle='--')
        ax.yaxis.grid(True, linestyle='--')

# Remove empty subplots
for i in range(len(states), n_rows * n_cols):
    row = i // n_cols
    col = i % n_cols
    fig.delaxes(axes[row][col])

# Adjust the layout
fig.tight_layout(pad=3)
plt.show()
```

```python
# What is the gender distribution of COVID-19 cases in each state and district? 

# Load the data into a pandas DataFrame
df = pat_data

# Remove rows where state or district is unassigned
df = df[(df['Detected State'] != 'Unassigned') & (df['Detected District'] != 'Unassigned')]

# Group the data by state, district, and gender and count the number of cases by gender
state_district_gender_counts = df.groupby(['Detected State', 'Detected District', 'Gender'])['Num Cases'].sum().reset_index()

# Get a list of unique states
states = state_district_gender_counts['Detected State'].unique()

# Set the number of rows and columns for the plot grid
n_rows = (len(states) + 1) // 2
n_cols = 2

# Create the plot grid
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(12, 6 * n_rows))

# Adjust the top margin of the plot grid
fig.subplots_adjust(top=0.95)

# Loop through each state and plot the gender distribution in its top 10 districts in a separate subplot
for i, state in enumerate(states):
    # Get the data for the current state and sort by number of cases
    state_data = state_district_gender_counts[state_district_gender_counts['Detected State'] == state].sort_values('Num Cases', ascending=False).head(10)
    if not state_data.empty:
        # Calculate the subplot position
        row = i // n_cols
        col = i % n_cols
        ax = axes[row][col]
        # Pivot the data to create a pie chart
        gender_counts = state_data.groupby('Gender')['Num Cases'].sum()
        # Create the pie chart in the current subplot
        gender_counts.plot(kind='pie', ax=ax, autopct='%1.1f%%', startangle=90, counterclock=False)
        ax.set_title(f'Gender Distribution of Cases in {state}')
        ax.set_ylabel('')
        ax.legend(title='Gender', loc='upper right', fontsize=8)
        ax.tick_params(axis='both', labelsize=8)

# Remove empty subplots
for i in range(len(states), n_rows * n_cols):
    row = i // n_cols
    col = i % n_cols
    fig.delaxes(axes[row][col])

# Adjust the layout
fig.tight_layout(pad=3)
plt.show()
```
