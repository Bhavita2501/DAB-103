# **Visualiztion of PAtient Data**

---



```python
import pandas as pd
import matplotlib.pyplot as plt


# select only necessary columns
age_bracket = pat_data[['Age Bracket', 'Num Cases']]

# group by age bracket and sum the number of cases
age_grouped = age_bracket.groupby(['Age Bracket']).sum()

# reset the index to make age bracket a column
age_grouped = age_grouped.reset_index()

# plot histogram
plt.hist(age_bracket['Age Bracket'], weights=age_bracket['Num Cases'], bins=10,color='skyblue',edgecolor='black', linewidth=1.2)
plt.xlabel('Age Bracket')
plt.ylabel('Number of Cases')
plt.title('Distribution of COVID-19 Cases by Age Bracket')
plt.show()
```
